```python
# Let's check about the null columns

null_cols = df.isnull().sum()
null_cols[null_cols > 0]
```

```python
# Dropping the columns with the highest number of null
drop_cols = list(null_cols[null_cols > 25000].index)
df = df.drop(drop_cols, axis=1)
```

```python
drop_cols
```

```python
df.info()
```

```python
# get rid of columns with reference or case number, kind of duplicated

cols = ['href formula', 'href', 'Case Number.1', 'Case Number.2']
tidy_sharks = df.drop(cols, axis=1)
```

```python
# Managing more missing data, change NaN by 'no info'

tidy_sharks.fillna('no info', inplace=True)
```

```python
# There are values up to index 6302

tidy_sharks.tail(-6300)
```

```python
# Eliminating indexes with no info

cut_sharks = tidy_sharks[:6302]
```
